```python
# 1. Understanding map with squaring of number
list_sample=[2,5,8,9,12]
list(map(lambda num : num*num,list_sample))
```

```python
# 2. Using map() to Convert Strings into Integers

str_num=['1','2','3','5']
list(map(lambda x:int(x),str_num))
```

```python
 # 3. Add corresponding elements of two lists using map().
 a=[12,2,3]
 b=[23,4,5]
 list(map(lambda x,y:x+y,a,b))
```

```python
 # 4. Use filter() to select only even numbers from a list.

 sample=[12,3,45,8,90]
 list(filter(lambda x : x%2==0,sample))
```

```python
# 5. Remove empty strings from a list using filter().

lst=["abc"," ","tabssum"," "]
list(filter(lambda st:len(st)>1,lst))
```

```python
 # 6. Problem Statement: Filter names whose length is greater than 5 characters.

 lst=["data structure","C programming","database","python","java"]
 list(filter(lambda word:len(word)>5,lst))
```

```python
# 7. Problem Statement: Use reduce() to find the sum of all numbers in a list.from functools import reduc
from functools import reduce
nums=[1,2,4,5,67,89,90]
reduce(lambda x,y:x+y,nums)
```

```python
# 8. Problem Statement: Use reduce() to multiply all numbers in a list.
lst=[1,2,3,4,5]
reduce(lambda x,y:x*y,lst)
```

```python
# 9. Problem Statement: Find the largest number from a list using reduce()

lst=[12,4,23,678,9,42]
reduce(lambda x,y:max(x,y),lst)
```

```python
# 10. Problem Statement: Select even numbers from a list and then square them.

lst=[1,2,4,5,6,8,99]
even=list(filter(lambda x:x%2==0, lst))
list(map(lambda x: x*x,even))
```

```python
# 10 . Problem Statement: Apply a normal function to every element of a list using map().

def square(x):
  return x*x

l=[1,2,3,4,5]
list(map(square,l))
```
