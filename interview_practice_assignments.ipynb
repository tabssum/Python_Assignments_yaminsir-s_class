```python
# Reverse the string by taking empty string and append char one by one

s="python programming"
reverse_str=""
for i in range(-1,-(len(s)+1),-1):
  reverse_str+=s[i]
print(reverse_str)
```

```python
# Create a dict of string where {index:char} using while loop

s="python"
sample_dict={}
i=0
while(i<len(s)):
  sample_dict[i]=s[i]
  i=i+1
print(f" dict is as shown: {sample_dict}")
```

```python
def create_index_char_dict(s):
  out={}
  for i in range(len(s)):
    out[i]=s[i]
  return out

print(create_index_char_dict("python"))
```

```python
#Armstrong number or not
num=int(input('Enter a number:'))
len_num=int(len(str(num)))
sum=0
for digit in str(num):
  sum=sum+int(digit)**len_num
print(sum)
if sum==num:
  print(f"{num} is armstrong nymber")
else:
  print("Not armstrong")
```
