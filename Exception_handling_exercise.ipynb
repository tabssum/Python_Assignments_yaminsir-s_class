```python
#  1. Handle Division by Zero
a=int(input("Enter first number :"))
b=int(input("Enter second number :"))
try:
  res=a/b
  print(f"Result is : {res}")
except ZeroDivisionError as msg:
  print(f"Exception occured : {msg}")
```

```python
#  2. Handle Invalid Input
a=input("Enter first number : ")
b=input("Enter second number :")
try:
  res=a*b
  print(f"Result of multiplcation is : {res}")
except Exception as msg:
  print(f"Cannot perform multiplication, Exception occured : {msg}")
```

```python
# 3. Write a program that handles ValueError and ZeroDivisionError separately.

try:
  a=int(input("Enter first number : "))
  b=int(input("Enter second number :"))
  res=a/b
  print(f"Result of division is : {res}")
except ZeroDivisionError as msg:
  print(f"Cannot perform divison, exception occured: {msg}")
except ValueError as msg:
  print(f"Cannot perform divison, value error exception occured, exception occured: {msg}")
```

```python
# 4.Write an example using try, except, else, and finally.

try:
  a=int(input("Enter a first number : "))
  print("User entered correct number")
except Exception as msg:
  print(f"Exception occured : {msg}")
else:
  print("Inside else block")
finally:
  print("Inside finally block")
```

```python
# 5. Write a Python program to handle list index out of range error.

list_sample=[1,2,3,5,67,8,9,45]
index=int(input("Enter a index to fetch element from list :"))
try:
  element = list_sample[index]
  print(f"Element found in the list at specified index : {element}")
except IndexError as msg:
  print(f"Exception occured : {msg}")
```

```python
# 7. Write a program that raises an exception if the number is negative.

num=int(input("Enter a number : "))
try:
  if num < 0:
    raise Exception("Number is less than zero")
except Exception as msg:
  print(msg)
```

```python
# 8. Custom Exception for Invalid Age

age=int(input("Enter your age : "))
try:
  if age < 18:
    raise Exception("Age is less than 18")
except Exception as msg:
  print(msg)
```

```python
# 9 . Safe File Opening


file_nm=input("Enter a file name : ")
try:
  f=open(file_nm,'r')
  print(f"Opened a file {file_nm}")
except Exception as msg:
  print(f"Exception occured : {msg}")
```

```python
# 10 . Write a program showing how finally is used for cleanup.

file_nm=input("Enter a file name : ")
try:
  f=open(file_nm,'r')
  print(f"Opened a file {file_nm}")
except Exception as msg:
  print(f"Exception occured : {msg}")
finally:
  print("Closing the file")
```
